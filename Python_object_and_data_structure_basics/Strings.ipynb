```python
'''
basic data types/data objects
int
float
str
basic data structures
list
dict
tup
set
bool
'''
# strings
print('\n**********strings**********')
s = 'hello world'

# built in function
print('\n**********built in function**********\n')
print(len(s))

# indexing
print('\n\n**********indexing**********\n')
print(s[-4:11])
print(s[::-1])
print(s[::11-1])
print(s[::-(11-1)])
print(s[::1-len(s)])

# cocatenate
print('\n\n**********cocatenate**********\n')
print(s+' of python')
print(s+(' and '+s)*2)
print(s[:5]+' python'+s[5:11])

# built in methods
print('\n\n**********built in methods**********\n')
print(s.upper())
print(s.lower())
print(s.split())
print(s.split('l'))

# print formatting with striings
print('\n\n**********print formatting with strings**********\n')
print(s+' of {} and {}'.format('python','linux'))
print(s+' of {1} and {0}.'.format('linux','python') )
print(s[:5]+' {} '.format('python')+s[6:])
name = 'python'
print(f'hello {name} world')
print(s[:5]+f' {name} '+s[6:])
```
