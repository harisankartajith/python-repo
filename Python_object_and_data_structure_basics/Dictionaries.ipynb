```python
print("********Dictionaries********\n")
'''
Unlike sequence, dictionaries works as mapping in python.
Unordered, there is no indexing and slicing.
'''
my_dictionary = {888:'Irah', 'key0':'Raknas'}
my_dictionary['key1'] = 634.3
my_dictionary['key2'] = [23, 5.2, 'legion']

print(f"Created dictionary is: {my_dictionary}; \nValues are: {my_dictionary['key1']}; {my_dictionary['key2'][2]}")
print(my_dictionary['key1'] - 34.3, my_dictionary['key2'][0] + my_dictionary['key2'][1])
print([x for x in my_dictionary['key2'][2]])
print([c for c in my_dictionary['key0'].upper() + my_dictionary[888].upper()])


print("\n\n********Nesting with dictionaries********\n")

# Nested dictionary

# Actually I think this is nested keys as dictionaries; i,e d['key'] and d['key']['nest_key'] are dictionaries
d = {'key':{'nest_key':{'subnest_key':'Ayner'}}}
print(f"Last value from dictionary d by keep calling the keys: {d['key']['nest_key']['subnest_key']}\n")

# This is the way I find to add nested_dictionaries in my_dictionary.
# where key3 and nestkey are dictionaries as well as keys
nestkey = {}
nestkey['subnestkey'] = 'Ayner'
key3 = {}
key3['nestkey'] = nestkey
my_dictionary['key3'] = key3
print(my_dictionary)

# key3 and nestkey are both keys and dictionaries and its values are key_value combination
print(f"Last value from my_dictionary by keep calling the keys: {my_dictionary['key3']['nestkey']['subnestkey']}")

# Showing nested_dictionaries
print(f"\nDictionary key3 is: {key3} \nDictionary nestkey is: {nestkey}")

# Showing values of keys 
print(f"Value of key3: {my_dictionary['key3']} \nValue of nestkey: {key3['nestkey']} \nValue of subnestkey: {nestkey['subnestkey']}")



print("\n\n********Dictionary methods********\n")

# keys()
print(f"Keys of my_dictionary: {my_dictionary.keys()}")
print(f"Its nested_dictionaries: {key3.keys()}; {nestkey.keys()}")
print(f"Keys of dictionary d: {d.keys()}; {d['key'].keys()}; {d['key']['nest_key'].keys()}")

# values()
print(f"\nValues of my_dictionary: {my_dictionary.values()}")
print(f"its nested_dictionaries: {key3.values()}; {nestkey.values()}")
print(f"Values of dictionary d: {d.values()}; {d['key'].values()}; {d['key']['nest_key'].values()}")

# items(): dict_items([(key, value), (key, value)])
print(f"\nItems of my_dictionary: {my_dictionary.items()}")
print(f"Its nested_dictionaries: {key3.items()}; {nestkey.items()}")
print(f"Items of dictionary d: {d.items()}; {d['key'].items()}; {d['key']['nest_key'].items()}")
```
