```python
import re

# Creating a dictionary of districts
districts = {1 : 'THIRUVANANTHAPURAM', 
             2 : 'KOLLAM', 
             3 : 'PATHANAMTHITTA', 
             4 : 'ALAPPUZHA', 
             5 : 'KOTTAYAM', 
             6 : 'IDUKKI', 
             7 : 'ERNAKULAM', 
             8 : 'THRISSUR', 
             9 : 'PALAKKAD', 
             10 : 'MALAPPURAM',
             11 : 'KOZHIKODE',
             12 : 'WAYANAD',
             13 : 'KANNUR',
             14 : 'KASARGOD'}



class TourGuide():
    
    def __init__(self):
        global districts
        print('\n')
        for i,j in (districts.items()):# To show the dictionary items
            print(i,':',j)
        
    def guidance(self, num):
        file = open('tour guide.txt')
        if num < 14:
            file.seek(0)
            a = re.search(districts[num], file.read())# Searching for district value of specified key.
            file.seek(0)
            b = re.search(districts[num+1], file.read())
            file.seek(0)
            print('\n',file.read()[a.start():b.start()])# Slicing with indexes a.start() and b.start().
        elif num == 14:
            file.seek(0)
            c = re.search(districts[num], file.read())# For getting the last district.
            file.seek(0)
            print('\n',file.read()[c.start():file.tell()])
        



def chatbot():
    print("TOURISM CHATBOT")
    print("---------------")
    print("Here are the top tourist attractions to see in Kerala.\n")
    
    while True:
        tg = TourGuide()
        
        try:
            num = int(input('\n Select your district :'))# For handling the error while inputing an alphabet.
        except ValueError:
            print("Invalid selection. Try once more.")
            continue
            
        if num in range(15):
            tg.guidance(num)
            cont1 = input("Do you want to try once more? :")# Whether to repeat it once more.
            if cont1.lower() in ['y', 'yes']:
                continue
            elif cont1.lower() in ['n', 'no']:# Program ending.
                print("Thank You!")
                break
            else:
                print("Invalid selection. Try once more :")
                continue
        else:
            print('Invalid selection.')# If the input for num is not in range of 1 to 14.
            cont = input("Do you want to continue? :")
            if cont.lower() in ['y', 'yes']:
                continue
            else:
                print("Thank You!")# Program ending.
                break
        break
```

```python
chatbot()
```
