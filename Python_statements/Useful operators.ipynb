```python
print("********Useful operators********\n")
'''
Cast to list is a method to present the result of some function, as in a list.
range function
enumerate function 
zip function
in operator
min and max functions
shuffle function and randint function from random library, which is a built-in library.
input function
'''

# range(including, upto i,e not inciuding, steps)
print(list(range(0,101,10)))# done casting to print

# enumerate()
mylist2 = []
print("\n\nfor loop with enumerate function\n")
for index_count, letter in enumerate('Raknas Irah'):
    print('{1} at index number {0}'.format(index_count, letter))
    mylist2.append(letter)
    
print(list(enumerate('Raknas Irah')))

# zip()
mylist1 = list(range(1,12))
print("\n\nmylist1: {}".format(mylist1))
print(f"mylist2: {mylist2}")
print(f"zip: {list(zip(mylist1, mylist2))}")# list of tuples

print("\nfor loop with zip()\n")

for e1,e2 in zip(mylist1, mylist2):
    if e1 == 1:
        print("{}'st letter is {}.".format(e1, e2))
    elif e1 == 2:
        print("{}'nd letter is {}.".format(e1, e2))
    elif e1 == 3:
        print("{}'rd letter is {}.".format(e1, e2))
    else:
        print("{}'th letter is {}.".format(e1, e2))

# in operator
print(f"\n\nSearch using 'in' operator: {'x' in ['x', 'y']}")
print('x' in [1,2,3])# this retuns False

# min and max functions
print(f"\n\nMinimum of mylist1 and mylist2: {min(mylist1)} and '{min(mylist2)}'")
print(f"Maximum of mylist1 and mylist2: {max(mylist1)} and '{max(mylist2)}'")

# random: built-in library
from random import shuffle, randint

#shuffle, won't reurn anything
shuffle(mylist1)
print(f'\n\nAfter shuffling list elements: {mylist1}')

# randint: return random integer in range [a, b], including both end points
print('\nrandint in a for loop\n')
for count in range(3):
    print(randint(0,100))

#input
name = input('\n\nEnter your name: ')
print(f"Thank you for reading my program {name}.")
```
